```python
import pandas as pd
from more import viz_helper as vh
```

```python
df = pd.read_csv("../data/titanic/train.csv")
df.info()
```

### Basic Plotting

**You want to see one side as white as possible and the other side as black as possible. This indicates maximum separation between the levels of the output (based on the features).**

```python
df_filtered = df[['Age','SibSp','Parch','Fare','Pclass','Survived']].dropna()
```

```python
%%time
vh.plot_similarity(data=df_filtered,y="Survived")
```

### Plotting for larger datasets

**Plotting for larger datasets may require sampling the data else it will take a long time to create these plots.**

```python
%%time
vh.plot_similarity(data=df_filtered,y="Survived",frac=0.3)
```
