```python
import pandas as pd
from more import viz_helper as vh
```

```python
df = pd.read_csv("../data/titanic/train.csv")
df.info()
```

```python
df_filtered = df[['Pclass','Age','SibSp','Parch','Fare']].dropna()
df_filtered.info()
```

### Running Basic PCA

```python
pcs = vh.perform_pca(data=df_filtered,n_components=3)
```

### Plotting PCs

```python
vh.plot_prin_comps(data=pcs)
```

### Running PCA and plotting coded scatterplot with target variable

```python
# Target variable "Survived" is added to the data.
df_filtered = df[['Pclass','Age','SibSp','Parch','Fare','Survived']].dropna()
df_filtered.info()
```

```python
# Target variable "Survived" is not used in calculating Principal Components but is used for plotting
pcs = vh.perform_pca(data=df_filtered,n_components=3,y='Survived')
vh.plot_prin_comps(data=pcs,hue='Survived')
```
