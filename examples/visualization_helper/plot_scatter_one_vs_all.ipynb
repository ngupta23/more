```python
import pandas as pd
from more import viz_helper as vh
```

```python
df = pd.read_csv("../data/titanic/train.csv")
df.info()
```

### Basic Plotting

```python
vh.plot_scatter_one_vs_all(data = df[['Age','Fare','SibSp','Parch','Embarked','Pclass']]
                           , y = 'Age'
                           , rows=1, cols=5
                           , alpha=0.5, figsize=(12,4))
```

### Coded Scatterplot

```python
vh.plot_scatter_one_vs_all(data = df[['Age','Fare','SibSp','Parch','Survived']]
                           , y = 'Age', hue = 'Survived'
                           , rows=1, cols=3
                           , alpha=0.5, figsize=(10,4))
```

### Autocalculation of rows and columns

```python
# Rows specified by user, Columns are autocalculated
vh.plot_scatter_one_vs_all(data = df[['Age','Fare','SibSp','Parch','Pclass']]
                           , y = 'Age', hue = 'Pclass'
                           , rows=1
                           , alpha=0.5, figsize=(10,4))
```

```python
# Columns specified by user, Rows are autocalculated
vh.plot_scatter_one_vs_all(data = df[['Age','Fare','SibSp','Parch','Sex']]
                           , y = 'Age', hue = 'Sex'
                           , rows=2
                           , alpha=0.5, figsize=(10,6))
```

```python
# Neither rows nor columns specified by user, hence both are autocalculated 
vh.plot_scatter_one_vs_all(data = df[['Age','Fare','SibSp','Parch','Sex']]
                           , y = 'Age', hue = 'Sex'
                           , alpha=0.5, figsize=(10,4))
```
